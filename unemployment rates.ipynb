```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['axes.facecolor'] = '#f1f1e6ff'
fig = plt.figure()
fig.set_facecolor('#f1f1e6ff')
```

```python
df = pd.read_excel("employment_sheet.xlsx")
ward_df = df[df['Ward '].isin([7, 8])]
print(ward_df)


df_7 = ward_df[ward_df["Ward "] == 7]
```

```python
# LABOR FORCE VS YEAR FOR WARD 7

avg_labor = df_7.groupby('Year ')['Labor Force '].mean()

# plot a line graph of the average labor force by word
fig = plt.figure()
fig.set_facecolor('#f1f1e6ff')
plt.plot(avg_labor.index, avg_labor.values, marker='o', color='#648c2eff')
plt.xlabel('Year')
plt.ylabel('Average Labor Force')
plt.title('Average Labor Force by Ward 7')
plt.show()
```

```python
# LABOR FORCE VS YEAR FOR WARD 8

df_8 = ward_df[ward_df["Ward "] == 8]
avg_labor = df_8.groupby('Year ')['Labor Force '].mean()

# plot a line graph of the average labor force by word
fig = plt.figure()
fig.set_facecolor('#f1f1e6ff')
plt.plot(avg_labor.index, avg_labor.values, marker='o', color = '#115880ff')
plt.xlabel('Year')
plt.ylabel('Average Labor Force')
plt.title('Average Labor Force by Ward 8')
plt.show()
```
